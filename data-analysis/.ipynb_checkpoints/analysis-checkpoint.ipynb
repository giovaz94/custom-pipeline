# Analysis of results

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

## Preparing data
We're going to prepare the data for the visualization on the plot. A csv file is produced by the monitor microservice; the data stored maintains:

    - The number of second passed (used as index)
    - The inbound workload
    - The Average latency
    - The dropped messages


```python
import pandas as pd

data = pd.read_csv("metrics.csv", index_col="Time(secs)")
data
```

# Filtering data
Filter some of the elements with inbound workload equal to 0

```python
filtered_data = data.iloc[900:1300]
```

## Inbound workload registered

```python
filtered_data["Inbound workload"].plot()
```

## Latency(ms)

```python
filtered_data.loc[filtered_data["Latency(ms)"] == 1_000_000] = 0
filtered_data["Latency(ms)"].plot()
```

## Loss

```python
filtered_data["Message loss"].plot()
```
