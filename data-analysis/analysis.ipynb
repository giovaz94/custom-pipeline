# Analysis of results

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import pandas as pd
```

## Preparing data
We're going to prepare the data for the visualization on the plot. A csv file is produced by the monitor microservice; the data stored maintains:

    - The number of second passed (used as index)
    - The inbound workload
    - The Average latency
    - The dropped messages


```python
global_scaling_folder = "global_scaling/"
local_scaling_folder = "local_scaling/"

gs_metrics = global_scaling_folder + "gs_metrics.csv"
ls_metrics = local_scaling_folder + "ls_metrics.csv"
```

```python
gs_df = pd.read_csv(gs_metrics)
ls_df = pd.read_csv(ls_metrics)
ls_df
```

```python
ls_df
```

## Latency

```python
gs_df.loc[gs_df["Latency(ms)"] == 1_000_000] = 0
ls_df.loc[ls_df["Latency(ms)"] == 1_000_000] = 0

plt.figure(figsize=(10, 6))
plt.plot(gs_df['Latency(ms)'], label='Latency - Global Scaling')
plt.plot(ls_df['Latency(ms)'], label='Latency - Local Scaling')
plt.legend()
```

# Message Loss

```python
plt.figure(figsize=(10, 6))
plt.plot(gs_df['Message loss'], label='Message loss - Global Scaling')
plt.plot(ls_df['Message loss'], label='Message loss - Local Scaling')
plt.legend()
```

# Deployed Instances

```python
plt.figure(figsize=(10, 6))
plt.plot(gs_df['Deployed Instances'], label='Deployed Instances - Global Scaling')
plt.plot(ls_df['Deployed Instances'], label='Deployed Instances - Local Scaling')
plt.legend()
```
